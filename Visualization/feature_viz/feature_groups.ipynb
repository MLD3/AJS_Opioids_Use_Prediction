```python
import pandas as pd
```

```python
import ast
```

```python
df = pd.read_csv('new_per_imp_data/no_bootstrap_correlated_group_imps_df_linear_svc_pu.csv')
```

```python
df
```

```python
df_names = pd.read_csv('../train_test/feature_names.csv')
```

```python
df['groups_list'] = df['groups'].apply(lambda l: ast.literal_eval(l))
```

```python
df
```

```python
df_names['group'] = 0
```

```python
df_names
```

```python
for i, l in enumerate(df['groups_list']):
    for v in l:
        df_names.loc[df_names['feature_name'] == v, 'group'] = i+1
```

```python
df_names
```

```python
df_names.to_csv('feature_groups.csv', index=False)
```
